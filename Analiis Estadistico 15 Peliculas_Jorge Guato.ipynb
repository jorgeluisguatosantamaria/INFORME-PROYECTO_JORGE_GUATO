```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

##Informe Estadistico de las primeras 15 peliculas en Rating

datos=pd.read_csv('PELICULAS1.csv')
df = pd.DataFrame(datos)
```

```python
print("Maxima duracion en las peliculas")
print(df['duration'].max())
minutos=df['duration'].max()
print("Horas:  ",minutos/60)
print("La pelicula que dura mas es: ")
print(df['duration'].idxmax())
##print(df.get_value(df['duration'].idxmax(),'movie_title'))
```

```python
print("Comparacion de likes entre generos")
df1=df.groupby(['genres'])[['movie_facebook_likes']].sum()
print(df1)
```

```python
print("El genero mas gustado en Facebook es:  ")
likes=df.groupby(['genres'])[['movie_facebook_likes']].sum()
print(likes['movie_facebook_likes'].idxmax(),end="")
```

```python
print("con: ",likes['movie_facebook_likes'].max()," likes")
```

```python
print("Presupuesto por director")
director=df.groupby(['director_name'])[['budget']].mean()
print(director)
```

```python
print("Genero con mas gasto")
dinero=df.groupby(['genres'])[['budget']].sum()
print("El monto es: ",dinero)
print("El genero: ",dinero['budget'].idxmax()," con",dinero['budget'].max())
```

```python
print("Actor Principal con mas ganancias")
actor=df.groupby(['actor_1_name'])[['gross']].sum()
print(actor['gross'].idxmax()," con",actor['gross'].max())
```

```python
df.groupby('genres')['movie_facebook_likes'].sum().plot(kind='bar',legend='Reverse')
plt.xlabel('Suma de likes en Facebook')
```

```python
df.movie_facebook_likes.groupby(df.genres).sum().plot(kind='pie',cmap='Paired')
plt.axis('equal')
plt.ylabel('')
plt.title("Promedio de likes en Facebook")
```

```python
sns.lmplot(x='num',y='movie_facebook_likes',data=df,fit_reg=False,hue='num',legend=False,palette='Paired')
```

```python
g=sns.lmplot(x='budget',y='imdb_score',data=df,palette='Set1')
```
